# Ingestion Exploration

Inspect local raw weather data before Airflow uploads it to MinIO.

```python
from pathlib import Path
import pandas as pd

raw_dir = Path('../data/raw/weather')
files = sorted(raw_dir.glob('*.csv'))
files
```

```python
df = pd.concat([pd.read_csv(path) for path in files], ignore_index=True)
df.head()
```

```python
df.assign(observed_at=pd.to_datetime(df['observed_at'])).describe(include='all')
```
