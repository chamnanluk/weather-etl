# Query Iceberg with Trino

Use the Trino Python client to query curated Iceberg tables.

```python
import pandas as pd
import trino

conn = trino.dbapi.connect(
    host='localhost',
    port=18080,
    user='trino',
    catalog='iceberg',
    schema='curated',
)

pd.read_sql('SELECT * FROM weather_daily_summary ORDER BY observed_date, station_id', conn)
```
